```python
import re
import numpy as np
import pandas as pd
from scipy.stats import kruskal

# -----------------------------
# 1. Read abundance table
# -----------------------------
abund = pd.read_csv('2025_Aug_new_analysis/data/Metagenomics_NextGen_Feature_on_Rows.txt', sep='\t')


abund = abund.dropna(subset=[abund.columns[0]])
abund = abund.rename(columns={abund.columns[0]: 'feature'})
abund.index = abund['feature'].astype(str)
abund = abund.drop(columns=['feature']).apply(pd.to_numeric, errors='coerce').fillna(0.0)

# calculate relative abundance
rel = abund.div(abund.sum(axis=0).replace(0, 1.0), axis=1)

# Shannon index per sample
shannon_values = []

for sample in rel.columns:
    p = rel[sample]
    p = p[p > 0]
    H = -(p * np.log(p)).sum()
    shannon_values.append(H)

shannon_values = np.array(shannon_values)

print(f"Mean Shannon ± SD: {shannon_values.mean():.3f} ± {shannon_values.std(ddof=1):.3f}")
print(f"N samples: {len(shannon_values)}")
```
